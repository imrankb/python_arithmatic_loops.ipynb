```python
# 1. Defining a Class
class Car:
    # Constructor (__init__) - initializes object attributes
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
    
    # Method (behavior of object)
    def display_info(self):
        print(f"{self.year} {self.brand} {self.model}")

# 2. Creating Objects (instances of the class)
car1 = Car("Toyota", "Corolla", 2020)
car2 = Car("Tesla", "Model S", 2023)

# 3. Using methods
car1.display_info()
car2.display_info()
```

```python
class ElectricCar(Car):   # ElectricCar inherits from Car
    def __init__(self, brand, model, year, battery_size):
        super().__init__(brand, model, year)  # reuse parent constructor
        self.battery_size = battery_size
    
    def display_info(self):
        print(f"{self.year} {self.brand} {self.model} - Battery: {self.battery_size} kWh")

# Object of child class
tesla = ElectricCar("Tesla", "Model 3", 2024, 75)
tesla.display_info()
```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance   # private attribute (name mangling)

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds!")

    def get_balance(self):
        return self.__balance

account = BankAccount("Ali", 1000)
account.deposit(500)
print(account.get_balance())   # 1500
```

```python
class MathTool:
    pi = 3.14159

    def __init__(self, radius):
        self.radius = radius

    # Instance method to calculate area        
    def area(self):
        return MathTool.pi * self.radius * self.radius

    # Class method to create a MathTool object from diameter
    @classmethod
    def from_diameter(cls, diameter):
        return cls(diameter / 2)

    # Static method to check if a value is a valid radius
    @staticmethod
    def is_valid_radius(radius):
        return radius > 0


# Example usage 
tool = MathTool(5)
print(tool.area())  # Area using radius

tool_from_diameter = MathTool.from_diameter(10)
print(tool_from_diameter.radius)  # Radius created from diameter

print(MathTool.is_valid_radius(-1))  # Static method check
```

```python
class Student:
    school_name = "City School"   # class attribute (shared by all students)

    def __init__(self, name, marks):
        self.name = name        # instance attribute
        self.marks = marks

    # Instance method → works with instance variables
    def display_info(self):
        return f"Name: {self.name}, Marks: {self.marks}"

    # Class method → works with the class itself
    @classmethod
    def change_school(cls, new_name):
        cls.school_name = new_name

    # Static method → utility method (does not depend on instance or class)
    @staticmethod
    def is_passing(marks):
        return marks >= 40


# Example usage
s1 = Student("Ali", 85)
s2 = Student("Sara", 32)

print(s1.display_info())   # Instance method
print(s2.display_info())

Student.change_school("Global Academy")   # Class method
print(Student.school_name)   # Shared by all students

print(Student.is_passing(85))  # Static method (True)
print(Student.is_passing(32))  # Static method (False)
```
