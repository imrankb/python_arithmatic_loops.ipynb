```python
class Vote:
    total_votes = {"PTI": 0, "PMLN": 0, "PPP": 0}

    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.has_voted = False

    def cast_vote(self, party):
        if self.age < 18:
            return f"{self.name}: Not eligible to vote!"

        if self.has_voted:
            return f"{self.name}: You already voted!"

        if party not in Vote.total_votes:
            return "Invalid party selected!"

        Vote.total_votes[party] += 1
        self.has_voted = True
        return f"{self.name} voted for {party}"

    @classmethod
    def show_results(cls):
        print("\n--- Election Results ---")
        for party, votes in cls.total_votes.items():
            print(f"{party}: {votes} votes")

        winner = max(cls.total_votes, key=cls.total_votes.get)
        print(f"\nWinner Party: {winner}")
v1 = Vote("Ali", 23)
v2 = Vote("Nasir", 17)
v3 = Vote("Javid", 30)

print(v1.cast_vote("PTI"))
print(v2.cast_vote("PMLN"))
print(v3.cast_vote("PPP"))
print(v1.cast_vote("PPP"))  # Trying to vote again

Vote.show_results()
```

```python
class Vote:
    total_votes = {"PTI": 0, "PMLN": 0, "PPP": 0}
    voted_cnic = set()   # store CNICs that already voted

    def __init__(self, name, age, cnic):
        self.name = name
        self.age = age
        self.cnic = cnic

    def cast_vote(self, party):
        if self.age < 18:
            return f"{self.name}: Not eligible to vote!"

        if self.cnic in Vote.voted_cnic:
            return f"{self.name}: This CNIC already voted!"

        if party not in Vote.total_votes:
            return "Invalid party selected!"

        Vote.total_votes[party] += 1
        Vote.voted_cnic.add(self.cnic)

        return f"{self.name} successfully voted for {party}"

    @classmethod
    def show_results(cls):
        print("\n--- Election Results ---")
        total = sum(cls.total_votes.values())

        for party, votes in cls.total_votes.items():
            percentage = (votes / total * 100) if total > 0 else 0
            print(f"{party}: {votes} votes ({percentage:.2f}%)")

        # Find winner or tie
        max_votes = max(cls.total_votes.values())
        winners = [party for party, votes in cls.total_votes.items() if votes == max_votes]

        if len(winners) > 1:
            print("\nResult: It's a tie between:", ", ".join(winners))
        else:
            print(f"\nWinner Party: {winners[0]}")
v1 = Vote("Ali", 23, "35202-1234567-1")
v2 = Vote("Nasir", 17, "35202-7654321-2")
v3 = Vote("Javid", 30, "35202-9999999-3")
v4 = Vote("Ali Again", 23, "35202-1234567-1")  # same CNIC

print(v1.cast_vote("PTI"))
print(v2.cast_vote("PMLN"))
print(v3.cast_vote("PPP"))
print(v4.cast_vote("PPP"))  # duplicate CNIC

Vote.show_results()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def details(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary

    def calculat_salary(self):
        print(f"Employee {self.name}'s  Salary: $ {self.salary}")

class Manager(Employee):
    def __init__(self, name,age, salary, bonus):
        super().__init__(name, age, salary)
        self.bonus = bonus

    def calculat_salary(self):
        total = self.salary + self.bonus 
        print(f"Manger {self.name}'s Salary: $ {total}") 

emp = Employee("Nadeem", 24, 18000) 
mang = Manager("Waseem", 35, 40000, 3000)

emp.details()
emp.calculat_salary()

mang.details()
mang.calculat_salary()
```

```python
import random

class BaseTeam:
    def __init__(self, name):
        self.name = name 
        self.played = 0
        self.points = 0
    def team_info(self):
        return f"{self.name} | P: {self.played} | Pts: {self.points}"

class CricketTeam(BaseTeam):
    def __init__(self, name):
        super().__init__(name)
        self.won = 0
        self.lost = 0
        self.runs_scored = 0
        self.runs_conceded = 0
    def play_match(self, runs_for, runs_against):
        self.played += 1
        self.runs_scored += runs_for
        self.runs_conceded += runs_against
        if runs_for > runs_against :
            self.won += 1
            self.points += 2
        if runs_for < runs_against:
            self.lost += 1
        else:
            self.points += 1

    def net_run_rate(self):
        if self.played == 0:
            return 0
        return (self.runs_scored) - (self.runs_conceded) / self.played
    
    def team_info(self):
        return (f"{self.name} |P: {self.played}"
               f"W:{self.won} |L: {self.lost}" 
               f"Pts: {self.points}"
               f"NRR: round{(self.net_run_rate(), 2)}")
    
class CricketLeague:
    def __init__(self, name):
        self.name = name
        self.teams = []

    def add_team(self, team):
        self.teams.append(team) 

    def simulate_match(self, team1, team2):
        score1 = random.randint(100, 250)
        score2 = random.randint(100, 250)
        team1.play_match(score1, score2)
        team2.play_match(score2, score1)

        print(f"{team1.name} ({score1}) Vs {team2.name} ({score2})")

    def simulate_tournament(self):
        print(f"\n-=- {self.name} Tournament Start!-=-\n")  
        for i in range(len(self.teams)):
            for j in range(i + 1 , len(self.teams)):
                self.simulate_match(self.teams[i], self.teams[j])
        self.points_table()
        self.announce_winner()

    def points_table(self):
        print("\n-=- Points Table -=-\n") 

        sorted_teams = sorted(self.teams, key = lambda t: (t.points, t.net_run_rate()), reverse = True)
        for team in sorted_teams:
            print(team.team_info()) 
        self.sorted_teams = sorted_teams

    def announmce_winner(self):
        winner = self.sorted_teams[0]
        print("\n-==- Champion: {winner.name}")                         

league = CricketLeague("NAAL PREMIER LEAGUE") 
team_names = [("BALOCH CLUB  "),("SANGAT BALOCH"),("BOLAN SPORTS "),("ZAKIR MAJEED "),
              ("NEARAN CLUB  "),("MHERAN TIGGAR"),("JALAWAN CLUB "),("SHAHEEED ARIF")] 

for name in team_names:
    league.add_team(CricketTeam(name))

league.simulate_tournament()      


```
